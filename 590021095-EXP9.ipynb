### Q1 & Q2: Student class, constructor, display, marks percentage, result, class average
Code Explanations : class average

class Student: : defines a class named Student

**init()** : constructor used to initialize student details

self.name = name : assigns name to the object

display() : displays all student details

marks_percentage() : calculates average marks of a student

result() : checks if marks in each subject are ≥ 40

students = [] : list to store student objects

Student(...) : creates student objects using constructor

total_percentage : stores sum of all students’ percentages

Class Average : calculates average percentage of the class

```python
class Student:
    def __init__(self, name, sap_id, phy, chem, maths):
        self.name = name
        self.sap_id = sap_id
        self.phy = phy
        self.chem = chem
        self.maths = maths

    def display(self):
        print("Name:", self.name)
        print("SAP ID:", self.sap_id)
        print("Physics:", self.phy)
        print("Chemistry:", self.chem)
        print("Maths:", self.maths)

    def marks_percentage(self):
        total = self.phy + self.chem + self.maths
        return total / 3

    def result(self):
        if self.phy >= 40 and self.chem >= 40 and self.maths >= 40:
            return "Pass"
        else:
            return "Fail"


students = []

for i in range(3):
    name = input("Enter name: ")
    sap_id = input("Enter SAP ID: ")
    phy = int(input("Enter Physics marks: "))
    chem = int(input("Enter Chemistry marks: "))
    maths = int(input("Enter Maths marks: "))
    students.append(Student(name, sap_id, phy, chem, maths))

total_percentage = 0

for s in students:
    s.display()
    print("Percentage:", s.marks_percentage())
    print("Result:", s.result())
    total_percentage += s.marks_percentage()
    print()

print("Class Average:", total_percentage / 3)
```

### Q3. Program to demonstrate Inheritance
Code Exaplanations: Inheritance

class Person: : parent (base) class

show_name() : method of parent class

class Student(Person): : child class inheriting Person

s = Student() : creates object of child class

s.show_name() : accesses parent class method

s.show_role() : accesses child class method

```python
class Person:
    def show_name(self):
        print("This is a person")

class Student(Person):
    def show_role(self):
        print("This is a student")

s = Student()
s.show_name()
s.show_role()
```

### Q4. Program to demonstrate Method Overriding
Code Explanations : Method Overidding

class Parent: : defines parent class

show() : method of parent class

class Child(Parent): : child class inherits Parent

show() : overrides parent class method

c.show() : calls overridden method of child class

```python
class Parent:
    def show(self):
        print("This is parent class method")

class Child(Parent):
    def show(self):
        print("This is child class method")

c = Child()
c.show()
```

### Q5: Operator Overloading (Adding two Point objects)
Code Explanations

class Point: : defines Point class

**init()** : initializes x and y values

**add()** : overloads + operator for Point objects

self.x + other.x : adds x values

self.y + other.y : adds y values

P1 + P2 : invokes overloaded + operator

P3.display() : displays resulting point values

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Point(self.x + other.x, self.y + other.y)

    def display(self):
        print("x =", self.x, "y =", self.y)


P1 = Point(10, 20)
P2 = Point(12, 15)
P3 = P1 + P2

P3.display()
```
