### 1. Write a Python function to find the maximum and minimum numbers from a sequence of numbers.
(Note: Do not use built-in functions.)

```python
def find_max_min(arr):
    max_val = arr[0]
    min_val = arr[0]

    for i in arr:
        if i > max_val:
            max_val = i
        if i < min_val:
            min_val = i

    return max_val, min_val

numbers = [10, 6, 8, 90, 12, 56]
print(find_max_min(numbers))
```

### 2. Write a Python function that takes a positive integer and returns the sum of the cube of all the positive integers smaller than the specified number.

```python
def sum_of_cubes(n):
    sum = 0
    for i in range(1, n):
        sum += i ** 3
    return sum

print(sum_of_cubes(5))
```

### 3. Write a Python function to print 1 to n using recursion.
(Note: Do not use loop.)

```python
def print_numbers(n):
    if n == 0:
        return
    print_numbers(n - 1)
    print(n)

print_numbers(5)
```

### 4. Write a recursive function to print Fibonacci series up to n terms.

```python
def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n - 1) + fibonacci(n - 2)

n = 7
for i in range(n):
    print(fibonacci(i), end=" ")
```

### 5. Write a lambda function to find volume of cone.

```python
volume_of_cone = lambda r, h: (1/3) * 3.14 * r * r * h

print(volume_of_cone(3, 7))
```

### 6. Write a lambda function which gives tuple of max and min from a list.
 `Sample Input: [10, 6, 8, 90, 12, 56]`

`Sample Output: (90, 6)`

```python
max_min = lambda arr: (
    max(arr),
    min(arr)
)

numbers = [10, 6, 8, 90, 12, 56]
print(max_min(numbers))
```

### Write functions to explain mentioned concepts:
a) Keyword Argument

```python
def student(name, age):
    print(name, age)

student(age=18, name="Anusha")
```

b) Default Argument

```python
def greet(name="Anusha"):
    print("Hello", name)

greet()
greet("Anusha")
```

c) Variable Length Argument

```python
def add_numbers(*numbers):
    sum = 0
    for i in numbers:
        sum += i
    print(sum)

add_numbers(1, 2, 3, 4, 5)
```
