## Lecture 1 — Strings

### 1) Initialization
* Strings are **immutable** sequences of characters.

* Create using:

   * Single quotes: `'hello'`
   * Double quotes: `"hello"`
   * Triple quotes (multi-line): `"""hello"""`

```python
s1 = 'Python'
s2 = "Data Science"
s3 = """Line1
Line2"""

print(s1)
print(s2)
print(s3)
```

### 2) Operators on strings

* `+` concatenation, `*` repetition
* `in` membership
* comparison uses lexicographic order

```python
print("Py" + "thon")      # 'Python'
print("ha" * 3)           # 'hahaha'
print("a" in "cat")       # True
print("abc" < "abd")      # True
```

```python
l =["Anusha", "Agarwal"]
"Anusha" in l
```

### 3) Indexing (0-based)

```python
s = "Python"
print(s[0])   # 'P'
print(s[-1])  # 'n'

print(s[len(s)-1])
# print(s[len(s)])
```

### 4) Slicing
Form: `s[start : stop : step]` (stop excluded)

```python
s = "Python"
print(s[1:4])    # 'yth'
print(s[:3])     # 'Pyt'
print(s[3:])     # 'hon'
print(s[5:1:-1])   # 'nohtyP'
```

### 5) `split()`
Splits into list by delimiter (default = whitespace)

```python
print("one two three".split())         # ['one','two','three']
print("2026-02-10".split("-")) 
```

## Lecture 2 — Lists
### 1) Initialization
Lists are **mutable**, ordered collections.

```python
L = [10, 20, 30]
mixed = [1, "A", 3.5, True]
empty = []
print(L)
print(mixed)
print(empty)
```

### 2) Common methods
* `append(x)`, `extend(iterable)`, `insert(i,x)`
* `remove(x)`, `pop(i)`, `clear()`
* `index(x)`, `count(x)`
* `sort()`, `reverse()`

```python
L = [3, 1, 2]
L.append(5)         # [3,1,2,5]
print(L)
L.sort()            # [1,2,3,5]
print(L)
L.pop(2)
print(L)
L.remove(5)
print(L)
L.clear()
print(L)
```

### 3) Operations
* Concatenation, repetition, membership
* indexing/slicing same as strings

```python
print([1,2] + [3,4])      # [1,2,3,4]
print([0]*4)              # [0,0,0,0]
print(2 in [1,2,3])       # True
```

### 4) List comprehension
Compact way to build lists.

```python
for item in iterator:
    return item*item
```

```python
squares = [x*x for x in range(1,6)]
evens = [x for x in range(10) if x%2==0]
pairs = [(x,y) for x in [1,2] for y in [3,4,6]]
print(squares)
print(evens)
print(pairs)
```

### 5) Nesting
List of lists (2D structure).

```python
mat = [[1,2,3],[4,5,6]]
print(mat[1][2])   # 6
```

## Lecture 3 — Tuples
### 1) Initialization
Tuples are **immutable** ordered collections.

```python
t1 = (1,2,3)
t2 = 1,2,3          # parentheses optional
t3 = (5,)           # single element tuple needs comma

print(t1)
print(t2)
print(t3)
```

### 2) Operations
* indexing/slicing, `+`, `*`, `in`

```python
t = (10,20,30,40)
print(t[1])      # 20
print(t[1:3])    # (20,30)
```

### 3) Methods (limited)

```python
print((1,2,2,3).count(2))  # 2
print((1,2,2,3).index(2)) 
```

### 4) Nesting

```python
T = ((1,2),(3,4))
print(T[1][0])  # 3
```

### 5) List vs Tuple


| Feature     | List            | Tuple                 |
|------------|-----------------|-----------------------|
| Mutability | Mutable         | Immutable             |
| Speed      | Slower          | Faster                |
| Methods    | Many            | Few                   |
| Use case   | Dynamic data    | Fixed records / keys  |

## Lecture 4 — Sets & Dictionaries
## A) Sets
### 1) Initialization
Sets are **unordered**, **unique** elements.

```python
S = {1,2,3}
S2 = set([1,2,2,3])   # {1,2,3}
empty_set = set()     # {} is empty dict, not set
print(S)
print(S2)
print(empty_set)
```

### 2) Methods / Operations
* `add`, `remove`, `discard`, `pop`
* `union(|)`, `intersection(&)`, `difference(-)`, `symmetric_difference(^)`

```python
A = {1,2,3}
B = {3,4,5}
print(A)
print(B)
print(A | B)   # {1,2,3,4,5}
print(A & B)   # {3}
print(A - B)   # {1,2}
```

## B) Dictionaries
### 1) Initialization
Key–value mapping. Keys must be hashable (immutable types).

```python
d = {"name":"Anusha", "age":18}
d2 = dict(a=1, b=2)
print(d)
print(d2)
```

### 2) Common operations/methods
* Access: `d[key]`, safe access: `d.get(key, default)`
* Insert/update: `d[key] = value`, `update()`
* Remove: `pop(key)`, `popitem()`, `del d[key]`
* Views: `keys()`, `values()`, `items()`

```python
print(d)
d["age"] = 19
print(d)
print(d.get("city", "NA"))    # 'NA'
```

### 3) Nesting

```python
student = {
  "name":"A",
  "marks":{"math":90, "python":95}
}
student["marks"]["python"]   # 95
```

### 4) Sorting & typecasting
* Sort dict by keys or by values using `sorted()`

```python
a="LKDJFHASDFJHSAD"
sorted(a)
```

```python
d = {"b":2, "a":5, "c":1}
print(sorted(d))                       # ['a','b','c']
print(sorted(d.items()))               # [('a',5),('b',2),('c',1)]
print(sorted(d.items(), key=lambda x: x[1]))  # by value
```

## Lecture 5 — Functions
### 1) User-defined functions

```python
def add(a, b):
    return a + b
```

### 2) Parameters
* positional, keyword, default, variable-length

```python
def f(a, b=10):
    return a+b

def g(*args):        # tuple of positional args
    return sum(args)

def h(**kwargs):     # dict of keyword args
    return kwargs
```

* Local, Enclosing, Global, Built-in (LEGB)

```python
x = 10
def foo():
    x = 5
    return x
```

### 5) Recursion
Function calls itself; must have base case.

```python
def fact(n):
    if n == 0:
        return 1
    return n * fact(n-1)
```

```python
print(fact(4))
```

## Lecture 6 — Advanced Functions
### 1) `lambda`
Anonymous function.

```python
print(square(2))
```

```python
def square(x):
    return x*x
```

```python
l =[1,2,3,4]
```

```python
def square(x):
    return x*x
new_list =[]
for each in l:
    new_list.append(square(each))
```

```python
[x*x for x in l]
list(map(lambda x : x*x, l))
```

```python
# A simple function to square a number
def square(number):
    return number * number

numbers = [1, 2, 3, 4, 5]

# Map the custom square function to the list
squared_numbers = map(square, numbers)

print(list(squared_numbers))
# Output: [1, 4, 9, 16, 25]
```

```python
list(map(square, l))
```

### 2) `map()`
Apply a function to every element.

```python
nums = [1,2,3]
list(map(lambda x: 100/x, nums))  # [1,4,9]
```

### 3) `filter()`
Keep only elements satisfying condition.

```python
nums = [1,2,3,4,5]
list(filter(lambda x: x%2==0, nums))  # [2,4]
```

### 4) Inner functions + returning functions

```python
def outer(msg):
    def inner():
        return msg.upper()
    return inner

f = outer("hello")
print(f)
print(f())   # 'HELLO'
```

```python
outer("hellp")()
```

### 5) Passing collections/functions

```python
def apply_twice(f, x):
    return f(f(x))

apply_twice(lambda a: a+1, 5)
```

## Lecture 7 — Mutability in functions + Applications + Doubts
### 1) Mutable vs Immutable behavior
**Immutable** types: `int, float, str, tuple` **Mutable** types: `list, dict, set`

Key idea:

* If you modify a mutable object inside a function, the caller sees it.
* If you “modify” immutable objects, you actually create a new object.

```python
def change_num(x):
    x = x + 1

def change_list(L):
    L.append(99)

a = 10
change_num(a)
print(a)     # 10 (unchanged)

L = [1,2]
change_list(L)
print(L)     # [1,2,99] (changed)
```
