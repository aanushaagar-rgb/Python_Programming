```python
n = 1111
reverse = 0 
temp = n
while temp !=0:
       reverse = (reverse * 10) + (temp % 10)
       temp = temp // 10
if n == reverse:
       print("Palindrome")
else:
       print("Not Palindrome")
              
```

```python
num = int(input("Enter a number: "))
temp = num
sum = 0
digits = len(str(num))

while temp > 0:
    digit = temp % 10
    sum += digit ** digits
    temp //= 10

if sum == num:
    print("Armstrong Number")
else:
    print("Not an Armstrong Number")
```
