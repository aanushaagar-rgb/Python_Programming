```python
def dummy_mileage_kmpl(engine_l: float, weight_kg: float, tyre_in: float) -> float:
    """
    Dummy (toy) mileage estimator in km/L.
    Inputs:
      engine_l : engine size in liters (e.g., 1.2, 1.5, 2.0)
      weight_kg: vehicle weight in kg (e.g., 900–2000)
      tyre_in  : tyre diameter in inches (e.g., 14–19)
    """
    mileage = 28 - 4 * engine_l - 0.006 * (weight_kg - 1000) - 0.7 * (tyre_in - 15)
    return max(5.0, mileage)  # clamp to minimum 5 km/L


def city_highway_kmpl(engine_l: float, weight_kg: float, tyre_in: float) -> tuple[float, float, float]:
    """
    Returns (base, city, highway) mileage in km/L using dummy multipliers.
    """
    base = dummy_mileage_kmpl(engine_l, weight_kg, tyre_in)
    city = 0.85 * base
    highway = 1.10 * base
    return base, city, highway


# Example usage:
if __name__ == "__main__":
    E, W, T = 1.5, 1200, 16
    base, city, highway = city_highway_kmpl(E, W, T)
    print(f"Inputs: engine={E}L, weight={W}kg, tyre={T}in")
    print(f"Base mileage   : {base:.2f} km/L")
    print(f"City mileage   : {city:.2f} km/L")
    print(f"Highway mileage: {highway:.2f} km/L")
```

```python
class Array:
    def __init__(self):
        self.arr = []

    
    def insert(self, value):
        self.arr.append(value)

    
    def insert_at(self, index, value):
        if 0 <= index <= len(self.arr):
            self.arr.insert(index, value)
        else:
            print("Index out of range")

    
    def delete(self, value):
        if value in self.arr:
            self.arr.remove(value)
        else:
            print("Value not found")

    
    def delete_at(self, index):
        if 0 <= index < len(self.arr):
            self.arr.pop(index)
        else:
            print("Index out of range")

    
    def update(self, index, value):
        if 0 <= index < len(self.arr):
            self.arr[index] = value
        else:
            print("Index out of range")

    
    def search(self, value):
        if value in self.arr:
            return self.arr.index(value)
        return -1

    
    def display(self):
        print(self.arr)

    
    def length(self):
        return len(self.arr)

    
    def sort(self):
        self.arr.sort()

    
    def reverse(self):
        self.arr.reverse()


    def maximum(self):
        if self.arr:
            return max(self.arr)
        return None

    
    def minimum(self):
        if self.arr:
            return min(self.arr)
        return None
```

```python
class Array:

    def __init__(self, arr):
        self.arr = arr

    def display(self):
        print("Array elements are:", self.arr)

    def find_sum(self):
        total = sum(self.arr)
        print("Sum of elements =", total)

    def find_max(self):
        maximum = max(self.arr)
        print("Maximum element =", maximum)

    def find_min(self):
        minimum = min(self.arr)
        print("Minimum element =", minimum)

    def search(self, key):
        if key in self.arr:
            print(key, "found in array")
        else:
            print(key, "not found in array")

    def sort_array(self):
        self.arr.sort()
        print("Sorted array:", self.arr)

    

    def matrix_add(self, A, B):
        result = []
        for i in range(len(A)):
            row = []
            for j in range(len(A[0])):
                row.append(A[i][j] + B[i][j])
            result.append(row)
        print("Matrix Addition Result:", result)

    def matrix_subtract(self, A, B):
        result = []
        for i in range(len(A)):
            row = []
            for j in range(len(A[0])):
                row.append(A[i][j] - B[i][j])
            result.append(row)
        print("Matrix Subtraction Result:", result)

    def matrix_multiply(self, A, B):
        result = []
        for i in range(len(A)):
            row = []
            for j in range(len(B[0])):
                sum_val = 0
                for k in range(len(B)):
                    sum_val += A[i][k] * B[k][j]
                row.append(sum_val)
            result.append(row)
        print("Matrix Multiplication Result:", result)

    def matrix_transpose(self, A):
        result = []
        for i in range(len(A[0])):
            row = []
            for j in range(len(A)):
                row.append(A[j][i])
            result.append(row)
        print("Matrix Transpose Result:", result)

    

    def dot_product(self, V1, V2):
        if len(V1) != len(V2):
            print("Dot product not possible! Vectors must be same length.")
            return

        dot = 0
        for i in range(len(V1)):
            dot += V1[i] * V2[i]

        print("Dot Product =", dot)

    def cross_product(self, V1, V2):
        if len(V1) != 3 or len(V2) != 3:
            print("Cross product only works for 3D vectors!")
            return

        cross = [
            V1[1] * V2[2] - V1[2] * V2[1],
            V1[2] * V2[0] - V1[0] * V2[2],
            V1[0] * V2[1] - V1[1] * V2[0]
        ]

        print("Cross Product =", cross)

    def matrix_range(self, A):
        maximum = A[0][0]
        minimum = A[0][0]

        for row in A:
            for val in row:
                if val > maximum:
                    maximum = val
                if val < minimum:
                    minimum = val

        print("Matrix Range (Max - Min) =", maximum - minimum)



a = Array([5, 2, 9, 1, 7])

a.display()
a.find_sum()
a.find_max()
a.find_min()
a.search(9)
a.sort_array()

M1 = [[1, 2],
      [3, 4]]

M2 = [[5, 6],
      [7, 8]]

print("\n--- Matrix Operations ---")
a.matrix_add(M1, M2)
a.matrix_subtract(M1, M2)
a.matrix_multiply(M1, M2)
a.matrix_transpose(M1)

print("\n--- Dot Product ---")
V1 = [1, 2, 3]
V2 = [4, 5, 6]
a.dot_product(V1, V2)

print("\n--- Cross Product ---")
a.cross_product(V1, V2)

print("\n--- Matrix Range ---")
a.matrix_range(M1)
```

```python
class Array:

    def __init__(self, arr):
        self.arr = arr

    

    def display(self):
        print("Array elements are:", self.arr)

    def find_sum(self):
        print("Sum of elements =", sum(self.arr))

    def find_max(self):
        print("Maximum element =", max(self.arr))

    def find_min(self):
        print("Minimum element =", min(self.arr))

    def search(self, key):
        if key in self.arr:
            print(key, "found in array")
        else:
            print(key, "not found in array")

    def sort_array(self):
        self.arr.sort()
        print("Sorted array:", self.arr)

    

    def is_valid_matrix(self, A):
        if not A:
            return False
        row_length = len(A[0])
        for row in A:
            if len(row) != row_length:
                return False
        return True

    def can_add_subtract(self, A, B):
        if not self.is_valid_matrix(A) or not self.is_valid_matrix(B):
            return False
        return len(A) == len(B) and len(A[0]) == len(B[0])

    def can_multiply(self, A, B):
        if not self.is_valid_matrix(A) or not self.is_valid_matrix(B):
            return False
        return len(A[0]) == len(B)

    def is_square(self, A):
        if not self.is_valid_matrix(A):
            return False
        return len(A) == len(A[0])

    def is_symmetric(self, A):
        if not self.is_square(A):
            return False
        for i in range(len(A)):
            for j in range(len(A)):
                if A[i][j] != A[j][i]:
                    return False
        return True

    

    def matrix_add(self, A, B):
        if not self.can_add_subtract(A, B):
            print("Matrix addition not possible! Dimensions must be same.")
            return
        result = []
        for i in range(len(A)):
            row = []
            for j in range(len(A[0])):
                row.append(A[i][j] + B[i][j])
            result.append(row)
        print("Matrix Addition Result:", result)

    def matrix_subtract(self, A, B):
        if not self.can_add_subtract(A, B):
            print("Matrix subtraction not possible! Dimensions must be same.")
            return
        result = []
        for i in range(len(A)):
            row = []
            for j in range(len(A[0])):
                row.append(A[i][j] - B[i][j])
            result.append(row)
        print("Matrix Subtraction Result:", result)

    def matrix_multiply(self, A, B):
        if not self.can_multiply(A, B):
            print("Matrix multiplication not possible! Columns of A must equal rows of B.")
            return
        result = []
        for i in range(len(A)):
            row = []
            for j in range(len(B[0])):
                sum_val = 0
                for k in range(len(B)):
                    sum_val += A[i][k] * B[k][j]
                row.append(sum_val)
            result.append(row)
        print("Matrix Multiplication Result:", result)

    def matrix_transpose(self, A):
        if not self.is_valid_matrix(A):
            print("Invalid matrix!")
            return
        result = []
        for i in range(len(A[0])):
            row = []
            for j in range(len(A)):
                row.append(A[j][i])
            result.append(row)
        print("Matrix Transpose Result:", result)

    def matrix_range(self, A):
        if not self.is_valid_matrix(A):
            print("Invalid matrix!")
            return
        maximum = A[0][0]
        minimum = A[0][0]
        for row in A:
            for val in row:
                if val > maximum:
                    maximum = val
                if val < minimum:
                    minimum = val
        print("Matrix Range (Max - Min) =", maximum - minimum)

    

    def dot_product(self, V1, V2):
        if len(V1) != len(V2):
            print("Dot product not possible! Vectors must be same length.")
            return
        dot = 0
        for i in range(len(V1)):
            dot += V1[i] * V2[i]
        print("Dot Product =", dot)

    def cross_product(self, V1, V2):
        if len(V1) != 3 or len(V2) != 3:
            print("Cross product only works for 3D vectors!")
            return
        cross = [
            V1[1] * V2[2] - V1[2] * V2[1],
            V1[2] * V2[0] - V1[0] * V2[2],
            V1[0] * V2[1] - V1[1] * V2[0]
        ]
        print("Cross Product =", cross)




a = Array([5, 2, 9, 1, 7])

print("----- ARRAY OPERATIONS -----")
a.display()
a.find_sum()
a.find_max()
a.find_min()
a.search(9)
a.sort_array()

M1 = [[1, 2],
      [3, 4]]

M2 = [[5, 6],
      [7, 8]]

print("\n----- MATRIX CHECKS -----")
print("Valid Matrix:", a.is_valid_matrix(M1))
print("Can Add:", a.can_add_subtract(M1, M2))
print("Can Multiply:", a.can_multiply(M1, M2))
print("Is Square:", a.is_square(M1))
print("Is Symmetric:", a.is_symmetric(M1))

print("\n----- MATRIX OPERATIONS -----")
a.matrix_add(M1, M2)
a.matrix_subtract(M1, M2)
a.matrix_multiply(M1, M2)
a.matrix_transpose(M1)
a.matrix_range(M1)

print("\n----- VECTOR OPERATIONS -----")
V1 = [1, 2, 3]
V2 = [4, 5, 6]
a.dot_product(V1, V2)
a.cross_product(V1, V2)

def replace_diagonal(self, A):
        if len(A) != len(A[0]):
            print("Diagonal replacement only possible for square matrix!")
            return
        print("Enter new diagonal values:")
        for i in range(len(A)):
            val = int(input(f"Enter value for position ({i},{i}): "))
            A[i][i] = val
        print("Matrix after replacing diagonal:")
        for row in A:
            print(row)

print("\n--- Replace Diagonal of Matrix 1 ---")
a.replace_diagonal(M1)
```
