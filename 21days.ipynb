## Find maximum in list

```python
numbers = [10,25,5,40]
max_value = numbers[0]

for i in numbers:
    if i > max_value:
        max_value = i

        print("Maximum=",max_value)
```

## Count vowels in a string

```python
string = "python"
count = 0

for ch in string:
    if ch in "aeiouAEIOU":
        count += 1

        print("Vowels=",count)
```

## Reverse a string

```python
string = "python"
rev = ""

for ch in string:
    rev = ch + rev
    print("Reversed=",rev)
```

## Prime number check

```python
num = 7
flag = True

if num <= 1:
    flag = False
else:
    for i in range(2, num):
        if num % i == 0:
            flag = False
            break

if flag:
    print("Prime Number")
else:
    print("Not prime")
```
