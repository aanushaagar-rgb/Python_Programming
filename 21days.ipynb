## Prime number check

```python
num = 7
flag = True

if num <= 1:
    flag = False
else:
    for i in range(2, num):
        if num % i == 0:
            flag = False
            break

if flag:
    print("Prime Number")
else:
    print("Not prime")
```

## Recursive Factorial

```python
def factorial(n):
    if n == 0:
        return 1
    return n * factorial(n-1)

print(factorial(5))
```

```python
num = int(input("Enter a number"))
temp = num
reverse = 0

while temp != 0:
    reverse = (reverse * 10) + (temp % 10)
    temp = temp // 10

if num == reverse:
        print("Palindrome")
else: 
        print("Not palindrome")
```

```python
num = int(input("Enter a number"))
temp = num
sum = 0
digits = len(str(num))

while temp > 0:
    digits = temp % 10
    sum += digits ** digits
    temp //= 10

if sum == num:
    print("Armstrong number")
else:
    print("Not a armstrong number")
```
